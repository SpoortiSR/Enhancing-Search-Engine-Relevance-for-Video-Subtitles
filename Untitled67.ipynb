```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer


lemmatizer = WordNetLemmatizer()

def preprocess(text):
    # Removing special characters and digits
    sentence = re.sub("[^a-zA-Z]", " ", text)

    # change sentence to lower case
    sentence = sentence.lower()

    # tokenize into words
    tokens = sentence.split()

    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    preprocessed_text = ' '.join(tokens)

    return preprocessed_text

```

```python

sampled_df['clean_file_content'] = sampled_df['file_content'].apply(preprocess)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
token_window = 500
overlap = 50

def chunk_document(document, token_window, overlap):
    words = document.split()
    chunks = []
    for i in range(0, len(words), token_window - overlap):
        chunk = ' '.join(words[i:i + token_window])
        chunks.append(chunk)
    return chunks

# Apply chunking to subtitle documents
sampled_df['Chunks'] = sampled_df['clean_file_content'].apply(lambda x: chunk_document(x, token_window, overlap))
```

```python
sampled_df.head()
```

```python
def encode_text(text):
    embedding=model.encode(text)
    return embedding
```

```python
sampled_df['embedding']=sampled_df['Chunks'].apply(encode_text)
```
